## Gráficos

Para realizar distintos tipos de gráficos se suele utilizar la librería *matplotlib.pyplot*.

```python
import matplotlib.pyplot as plt
```

**Line Plots (Gráfico de Líneas):**

Para este ejemplo crearemos dos listas.

```python
x = [2016,2017,2018,2019,2020,2021]
y = [45,46,47,50,54,60]
plt.plot(x,y)
plt.show()
```

Estas listas deben tener ambas las mismas dimensiones, y ademas se agrega el comando *plt.show()* para evitar que aparezcan textos adicionales.

Podemos personalizar más los gráficos a partir de agregar parámetros en plt.plot, nombre de los ejes,cambiar el tamaño de la figura, cantidad de ticks por eje, etc.

```python
plt.figure(figsize=(10,6))#Cambiar el tamaño del plot
plt.plot(x,y, marker='o',linestyle='--',color='r')
plt.xlabel("Años")
plt.ylabel("Poblacion (M)")
plt.yticks([45,50,55,60]) #Se utiliza para no mostrar todos los valores del eje
plt.show()
```

Para el caso que querramos graficar dos listas en un mismo plot para realizar una comparación, debemos tener la lista con los datos de ambos ejes para cada gráfico, y realizar dos veces *plt.plot*, una para cada gráfico. 

Para esto, podemos agregar una leyenda, para saber que representa cada línea. Debemos agregar el nombre del gráfico, esto se hace agregando *label=nombre_grafico* en el comando plt.plot.
Luego, con el comando *plt.legend()* agregamos la misma, opcionalmente se puede agregar en este comando *loc="ubicacion_leyenda"* para modificar la posición de la leyenda.

```python
plt.figure(figsize=(10,6))

#País 1
x1 = [2016,2017,2018,2019,2020,2021]
y1 = [45,46,47,50,54,60]

#País 2
x2 = [2016,2017,2018,2019,2020,2021]
y2 = [47,46,45,48,50,53]

plt.plot(x1,y1,marker="o",linestyle="--",color="g", label="Colombia")
plt.plot(x2,y2,marker="d",linestyle="-",color="b", label="Argentina")
plt.legend(loc="lower right")
plt.savefig("./plots/plot1.png")
plt.show()
```

Podemos guardar estos gráficos como imagenes, para esto utilizamos el siguiente comando *plt.savefig("ruta_y_nombre_archivo.formato")*, especificando la ruta en donde se va a guardar, el nombre del archivo, formato y tamaño. Siempre el comando debe estar dentro de la misma celda en la cual definimos el gráfico. En lo posible realizar todo lo correspondiente al gráfico en la misma celda.

## Subplots

Los subplots es una forma de tener varios gráficos uno al lado del otro.


El comando para crear subplots es *fig, ax = plt.subplots( filas, columnas)* de donde *ax* es una lista con los ejes de coordenadas de cada plot creado.

```python
fig, ax = plt.subplots(1,2, sharey=True) #El último comando es para que compartan el eje Y

ax[0].plot(x1,y1, color="g",label="Colombia")
ax[0].legend()
ax[1].plot(x2,y2, color="r", label="Argentina")
ax[1].legend()
plt.show()
```

## Bar Plots (Gráfico de barras)

Lo correspondiente a agregados de gráficos, como nombre de ejes, títulos, y demás parámetros se realizan de la misma manera en todos los tipos de Plots. Es por esto que solo realizaremos unos simples ejemplos para conocer los restantes tipos de gráficos.

El comando para realizar gráfico de barras es *plt.bar(x,y)*

```python
x = ["Argentina","Colombia","Perú"]
y = [40,50,33]
plt.bar(x,y)
plt.show()
```

## Piecharts (gráfico de torta)

El comando para realizar gráfico de tortas es *plt.pie (dato_numerico, labels=nombre_porcion )*, y como se especifica, primero debemos ingresar el dato numérico, y luego el nombre de cada porción como label.

```python
plt.pie(y, labels=x)
plt.show()
```

## Histogramas

Para realizar histogramas debemos definir el ancho de las barras como *bins*.

```python
edades = [10,15,16,18,20,22,23,24,25,26,29,35,50,55,61]
bins = [10,15,20,25,30,35,40,45,50,55,60,65]

plt.hist(edades,bins,edgecolor="black") #Agregamos un reborde negro a las barras
plt.show()
```

## Boxplots

Este tipo de gráficos se utiliza para graficar su valor medio en el centro de la caja, y sus cuartiles en los extremos. Sin embargo es de mucha utilidad debido a que los valores out of the box son eventos "atípicos" que se podrían estar buscando.

El comando para realizar el gráfico de cajas es *plt.boxplot(lista)*

```python
edades = [10,15,16,18,20,22,23,24,25,26,29,35,60,61]
plt.boxplot(edades)
plt.show()
```

## Scatterplot (Gráfico de dispersión)

Se utiliza mucho para buscar si dos variables están correlacionadas.

```python
a = [1,5,3,4,6,5,5,8]
b = [2,2,3,6,8,9,5,7]
plt.scatter(a,b)
plt.show()
```

**Ejemplo:**

Para crear un gráfico mediante un data frame debemos especificar dentro de este una lista, y los nombres de las columnas.

```python
import pandas as pd
#Crear listas
bj=["Boca Juniors",18,1905]
rp=["River Plate",12,1901]
ind=["Independiente",18,1905]

#Creamos una lista de las listas
lista_eq = [bj,rp,ind]

#Creamos un df a partir de las listas
df_eq = pd.DataFrame(lista_eq,columns = ["nombre","titulos","fundacion"])
df_eq
```

```python
#Grafico de torta
plt.pie(df_eq["titulos"],
        labels = df_eq["nombre"], #para colocar nombre a las porciones
        colors = ["yellow","red","green"],#para definir los colores de las porciones
        autopct = '%1.1f%%')#para mostrar el porcentaje de la porcion en la torta
plt.show()
```

```python
#Grafico de barras
plt.bar(df_eq["nombre"], #eje x
        df_eq["titulos"], #eje y
        color=["yellow","red","green"])
plt.title("Títulos internacionales")
plt.yticks([0,4,8,12,14,16,18])
plt.savefig("./plots/Clubes.png",dpi=500,facecolor='w') #dpi es la resolución 
```
