## Data Frame ## 

Es una "tabla", por lo que está formado por filas y columnas.

Cada **columna** representa una **variable diferente** (por ejemplo edad).

Cada **fila** representa una **observación diferente** (por ejemplo una persona).

Los elementos de cada columna deben ser del **mismo tipo**. Por ejemplo, todos los elementos de la columna A deben ser numéricos y todos los elementos de la columna B deben ser strings.

Las columnas deben tener el **mismo número de elementos**. Por ejemplo, la columna A debe teer 6 elementos, y la columna B también debe tener 6 elementos.

Los datos de un Data Frame pueden ser modificados.

Para definir un Data Frame utilizaremos el móduo *pandas*.

```python
#importar pandas
import pandas as pd
```

```python
# 1) Creamos un data frame a partir de un diccionario, primero definimos el diccionario
```

```python
datos = {"a": [0,1,2,3,5],"b":["a","v","s","f","w"],"c":[10,20,30,40,50] }
datos
```

```python
#Creamos el data frame a partir de ese diccionario (casi siempre se lo denomina df)
df = pd.DataFrame(data = datos)
df
```

También podemos crear un data frame a partir de varias listas.

# Cada lista será una fila del data frame

l1 = [1,2,3]
l2 = [4,5,6]


```python
# Creamos el data frame, debemos asignar un nombre a las columnas 
df = pd.DataFrame(data = [l1,l2], columns = ["c1","c2","c3"])
df
```

## Filtrado ##

Por un lado podemos filtrar y quitar estas filas que contienen datos NaN mediante el comando *df.dopna()*.

```python
df_filtrado = df.dropna()
df_filtrado.head()
```

Otra alternativa es llenar esos espacios con un valor establecido mediante el comando *df.fillna(0)*.

```python
df_filtrado=df.fillna(0)
df_filtrado.head()
```

También puedo llenar datos dependiendo de la columna, esto se realiza mediante un diccionario en el argumento de *df.fillna()*.

```python
df_filtrado = df.fillna({"retweets":0,"mentions":-1})
df_filtrado.head()
```

Para poder visualizar solamente los datos de una columna dada podemos realizarlo de la siguiente manera:


```python
df["full_text"]
```

Para visualizar varias columnas a la vez deberemos ingresar al argumento de df[] una lista con todas las columnas deseadas.

Para discriminar entre filas podemos hacerlo entre los índices (0,1,2,...), entre el identificador (id), o mediante condiciones.

Para obtener la fila mediante su índice, utilizamos la función *df_iloc[ ]*, puede ser una sola fila indicandole un solo valor en el argumento, o pueden ser varias indicandolas mediante a:b, también pueden ser filas de índices aislados utilizando una lista con sus índices.

Para poder filtrar por otro parámetro, por ejemplo "id", utilizamos la función *df.loc[ ]*

```python
df.loc[[183723,183737]]
```

Mediante la función *df.loc[ ]* también puedo filtrar por filas y columnas a la vez, en el argumento ponemos una lista que me represente las filas que quiero, y otra lista que me represente las columnas.

```python
df.loc[[183721,183722], ["favorites", "full_text"]]
```

Uno de los filtrados más importante es el por condiciones, en donde quiero filtrar filas en donde en alguna de sus columnas se cumpla una o unas condiciones específicas.

Supongamos que queremos filtrar para obtener todas las filas cuyos favoritos superen los 400. Esto se realiza con la siguiente sintaxis *df[df["columna"]**condición** ]*.

```python
df2=df
df2[df2["favorites"]>400]
```

Para concatenar condiciones se realiza con esta sintaxis *(condición_1) & (condición_2)*, así con el símbolo *&* es para una condición **and** y | es para una condición **or** :

```python
df3=df
df3=df3[(df3["favorites"]>400) & (df2["mentions"] > 20)]
df3
```

Supongamos ahora que queremos obtener todas las filas en donde el texto del tweet,o alguna otra columna, tenga cierta palabra (string), para esto utilizamos el siguiente comando *df["nombre_columna"].str.contains("palabra_buscada")*

```python
df4=df
df4[df4["full_text"].str.contains("programming")]
```

Una transformación de datos es realizar operaciones o algoritmos con columnas existentes para generar nuevas columnas con estos datos calculados, por ejemplo podemos dar un ejemplo de "popularidad", que sería la relación entre las personas que te siguen (Followers) vs las personas que seguís (Followees). Para esto primero debemos definir la función de transformación.


```python
#Definimos la nueva función
def popularidad(fila):
    resultado = fila["followers"]/fila["followees"]
    return resultado

#Recien ahora aplicamos la función para crear la nueva columna
df5=df
df5["popularidad"] = df5.apply(popularidad, axis=1) 
df5[["full_text","popularidad"]]

```


Supongamos que queremos realizar un agrupamiento según algún parámetro, por ejemplo el país. Agrupar significa unir filas que tengan un mismo valor dentro de una columna. Estos datos sufren una transformación, por ejemplo el proedio, obtener el máximo valor, el mínimo, etc. Esto lo realizamos mediante la función *df.groupby("columna").función()*, esta función será *mean()* para obtener un promedio.


```python
df.groupby("country").mean()
```

Como vemos, ahora el identificador pasa a ser el país, y los valores de las columnas se expresan en su valor promedio.

También puedo aplicar diferentes funciónes a diferentes columnas mediante el uso *.agg({diccionario})* y de un diccionario.


```python
grouped = df.groupby("country").agg({
"followers":'sum',
"mentions":'mean',
"retweets":'max'})
grouped
```

## Graficación de Datos ## 

Para poder graficar datos debemos hacer uso de la librería *matplotlib* . 

Supongamos que queremos graficar la cantidad de seguidores según el país, para esto podemos visualizar un grafico de barras (por defecto es de líneas) mediante el comando:


```python
import matplotlib.pyplot as plt
grouped["followers"].plot(kind = "bar")

plt.show()
```

Para observar si existe algun tipo de correlación entre dos variables, por ejemplo entre seguidores y retweets, podemos realizar un gráfico de dispersión de puntos "scatter" colocando los valores del "eje x" y "eje y" en el comando.

```python
df.plot(kind="scatter", x="favorites", y="retweets")
plt.show()
```

## Exportar Data Frame ##

Una vez que hayamos hecho un filtrado o alguna modificación a un data frame siempre vamos a querer exportarlo para volver a tener un archivo del mismo caracter que se importo en su momento. Para esto utilizamos el siguiente comando que me creará el archivo solicitado en la carpeta en donde nos encontremos. *nombre_df.to_tipo_de_archivo("nombre_archivo.tipo_archivo")*

```python
grouped.to_csv("salida.csv")
```

En la carpeta en la que nos encontramos debería aparecer el archivo "salida.csv".

Así como existe to_csv, también podría ser to_json, to_html, to_xls, to_sql, etc.
