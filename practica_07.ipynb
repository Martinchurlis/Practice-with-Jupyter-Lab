## Data Frame ## 

Es una "tabla", por lo que est√° formado por filas y columnas.

Cada **columna** representa una **variable diferente** (por ejemplo edad).

Cada **fila** representa una **observaci√≥n diferente** (por ejemplo una persona).

Los elementos de cada columna deben ser del **mismo tipo**. Por ejemplo, todos los elementos de la columna A deben ser num√©ricos y todos los elementos de la columna B deben ser strings.

Las columnas deben tener el **mismo n√∫mero de elementos**. Por ejemplo, la columna A debe teer 6 elementos, y la columna B tambi√©n debe tener 6 elementos.

Los datos de un Data Frame pueden ser modificados.

Para definir un Data Frame utilizaremos el m√≥duo *pandas*.

```python
#importar pandas
import pandas as pd
```

```python
# 1) Creamos un data frame a partir de un diccionario, primero definimos el diccionario
```

```python
datos = {"a": [0,1,2,3,5],"b":["a","v","s","f","w"],"c":[10,20,30,40,50] }
datos
```

```python
#Creamos el data frame a partir de ese diccionario (casi siempre se lo denomina df)
df = pd.DataFrame(data = datos)
df
```

Tambi√©n podemos crear un data frame a partir de varias listas.

# Cada lista ser√° una fila del data frame

l1 = [1,2,3]
l2 = [4,5,6]


```python
# Creamos el data frame, debemos asignar un nombre a las columnas 
df = pd.DataFrame(data = [l1,l2], columns = ["c1","c2","c3"])
df
```

## Filtrado ##

Mediante la funci√≥n *df.loc[ ]* tambi√©n puedo filtrar por filas y columnas a la vez, en el argumento ponemos una lista que me represente las filas que quiero, y otra lista que me represente las columnas.

```python
df.loc[[183721,183722], ["favorites", "full_text"]]
```

Uno de los filtrados m√°s importante es el por condiciones, en donde quiero filtrar filas en donde en alguna de sus columnas se cumpla una o unas condiciones espec√≠ficas.

Supongamos que queremos filtrar para obtener todas las filas cuyos favoritos superen los 400. Esto se realiza con la siguiente sintaxis *df[df["columna"]**condici√≥n** ]*.

```python
df2=df
df2[df2["favorites"]>400]
```

Para concatenar condiciones se realiza con esta sintaxis *(condici√≥n_1) & (condici√≥n_2)*, as√≠ con el s√≠mbolo *&* es para una condici√≥n **and** y | es para una condici√≥n **or** :

```python
df3=df
df3=df3[(df3["favorites"]>400) & (df2["mentions"] > 20)]
df3
```

Supongamos ahora que queremos obtener todas las filas en donde el texto del tweet,o alguna otra columna, tenga cierta palabra (string), para esto utilizamos el siguiente comando *df["nombre_columna"].str.contains("palabra_buscada")*

```python
df4=df
df4[df4["full_text"].str.contains("programming")]
```

Una transformaci√≥n de datos es realizar operaciones o algoritmos con columnas existentes para generar nuevas columnas con estos datos calculados, por ejemplo podemos dar un ejemplo de "popularidad", que ser√≠a la relaci√≥n entre las personas que te siguen (Followers) vs las personas que segu√≠s (Followees). Para esto primero debemos definir la funci√≥n de transformaci√≥n.


```python
#Definimos la nueva funci√≥n
def popularidad(fila):
    resultado = fila["followers"]/fila["followees"]
    return resultado

#Recien ahora aplicamos la funci√≥n para crear la nueva columna
df5=df
df5["popularidad"] = df5.apply(popularidad, axis=1) 
df5[["full_text","popularidad"]]

```


Supongamos que queremos realizar un agrupamiento seg√∫n alg√∫n par√°metro, por ejemplo el pa√≠s. Agrupar significa unir filas que tengan un mismo valor dentro de una columna. Estos datos sufren una transformaci√≥n, por ejemplo el proedio, obtener el m√°ximo valor, el m√≠nimo, etc. Esto lo realizamos mediante la funci√≥n *df.groupby("columna").funci√≥n()*, esta funci√≥n ser√° *mean()* para obtener un promedio.


```python
df.groupby("country").mean()
```

Como vemos, ahora el identificador pasa a ser el pa√≠s, y los valores de las columnas se expresan en su valor promedio.

Tambi√©n puedo aplicar diferentes funci√≥nes a diferentes columnas mediante el uso *.agg({diccionario})* y de un diccionario.


```python
grouped = df.groupby("country").agg({
"followers":'sum',
"mentions":'mean',
"retweets":'max'})
grouped
```

## Graficaci√≥n de Datos ## 

Para poder graficar datos debemos hacer uso de la librer√≠a *matplotlib* . 

Supongamos que queremos graficar la cantidad de seguidores seg√∫n el pa√≠s, para esto podemos visualizar un grafico de barras (por defecto es de l√≠neas) mediante el comando:


```python
import matplotlib.pyplot as plt
grouped["followers"].plot(kind = "bar")

plt.show()
```

Para observar si existe algun tipo de correlaci√≥n entre dos variables, por ejemplo entre seguidores y retweets, podemos realizar un gr√°fico de dispersi√≥n de puntos "scatter" colocando los valores del "eje x" y "eje y" en el comando.

```python
df.plot(kind="scatter", x="favorites", y="retweets")
plt.show()
```

## Exportar Data Frame ##

Una vez que hayamos hecho un filtrado o alguna modificaci√≥n a un data frame siempre vamos a querer exportarlo para volver a tener un archivo del mismo caracter que se importo en su momento. Para esto utilizamos el siguiente comando que me crear√° el archivo solicitado en la carpeta en donde nos encontremos. *nombre_df.to_tipo_de_archivo("nombre_archivo.tipo_archivo")*

```python
grouped.to_csv("salida.csv")
```

En la carpeta en la que nos encontramos deber√≠a aparecer el archivo "salida.csv".
